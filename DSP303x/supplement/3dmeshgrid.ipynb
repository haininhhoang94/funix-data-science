```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define ranges and number of points for x, y axes
min_val, max_val = -2, 2
num_points = 100

# Create coordinate vectors for x, y axes
x_values = np.linspace(min_val, max_val, num_points)
y_values = np.linspace(min_val, max_val, num_points)
```

```python
# Generate meshgrid for x, y axes
X, Y = np.meshgrid(x_values, y_values)

# Define a function to create a mesh plane (example: a plane in z = 0)
Z = np.zeros_like(X)  # Z-values for a plane (set to zero for z = 0)
```

```python

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the mesh plane
ax.plot_surface(X, Y, Z, cmap="viridis")

# Set labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("Mesh Plane in 3D")

plt.show()
```
