```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
def model(x, a, b):
    return a * x + b
```

```python
# True parameters
a_true = 2
b_true = 5

# Generate some data points with noise
xdata = np.linspace(0, 10, 100)
ydata = model(xdata, a_true, b_true) + np.random.normal(size=xdata.size)
```

```python
# Fit the model to the data
popt, pcov = curve_fit(model, xdata, ydata)

# Extracting the estimated parameters
a_est, b_est = popt
print(f"Estimated Parameters: a = {a_est}, b = {b_est}")
```

```python
# Plot the original data and the fitted line
plt.scatter(xdata, ydata, label='Data')
plt.plot(xdata, model(xdata, *popt), color='red', label='Fitted line')
plt.legend()
plt.show()
```
