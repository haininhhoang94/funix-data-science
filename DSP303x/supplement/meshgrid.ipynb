```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define ranges and number of points
min_x, max_x = -2, 2
min_y, max_y = -3, 3
num_points_x, num_points_y = 100, 150
```

```python
# Create coordinate vectors
x_values = np.linspace(min_x, max_x, num_points_x)
y_values = np.linspace(min_y, max_y, num_points_y)

# Generate meshgrid
X, Y = np.meshgrid(x_values, y_values)

# Plotting the grid points
plt.scatter(X, Y, s=5)  # Scatter plot of grid points
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Meshgrid of Coordinates')
plt.show()
```
