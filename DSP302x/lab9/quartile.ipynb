```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
data = np.random.randn(1000)  # 100 random numbers from a normal distribution

# Create a DataFrame from the data
df = pd.DataFrame(data, columns=["Data"])

# Calculate the quartiles
Q1 = df["Data"].quantile(0.25)
Q2 = df["Data"].quantile(0.5)
Q3 = df["Data"].quantile(0.75)

# Create a histogram
plt.hist(df["Data"], bins=10, color="blue", alpha=0.7)

# Overlay the quartiles on the histogram
plt.axvline(Q1, color="red", linestyle="dashed", linewidth=2, label="Q1")
plt.axvline(Q2, color="green", linestyle="dashed", linewidth=2, label="Q2 (Median)")
plt.axvline(Q3, color="yellow", linestyle="dashed", linewidth=2, label="Q3")

# Add a legend
plt.legend()

# Add title and labels
plt.title("Histogram with Quartiles")
plt.xlabel("Data")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Generate some data
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

# Create a DataFrame from the data
df = pd.DataFrame(data, columns=["Data"])

# Calculate the quartiles
Q1 = df["Data"].quantile(0.25)
Q2 = df["Data"].quantile(0.5)
Q3 = df["Data"].quantile(0.75)

# Create a cumulative distribution function (CDF) plot
sns.ecdfplot(df["Data"], color="blue", label='CDF')

# Overlay the quartiles on the CDF plot
plt.axvline(Q1, color="red", linestyle="dashed", linewidth=2, label="Q1")
plt.axvline(Q2, color="green", linestyle="dashed", linewidth=2, label="Q2 (Median)")
plt.axvline(Q3, color="yellow", linestyle="dashed", linewidth=2, label="Q3")

# Add a legend
plt.legend()

# Add title and labels
plt.title("CDF with Quartiles")
plt.xlabel("Data")
plt.ylabel("Cumulative Probability")

# Show the plot
plt.show()
```
